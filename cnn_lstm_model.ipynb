### Load in the streamwater data

```python
import numpy as np
import pandas as pd
```

```python
streamwater_data = pd.read_csv("data/streamflow_data/Final_Processed_Station_Data_Watershed.csv")
streamwater_data.dropna(inplace=True)
streamwater_data
```

```python
station_lats = streamwater_data.iloc[0][1:]
station_lons = streamwater_data.iloc[1][1:]
watersheds = streamwater_data.iloc[2][1:]

# drop the first 3 rows
streamwater_data = streamwater_data.drop([0, 1, 2])
# drop the first column
streamwater_data = streamwater_data.drop(columns=["name"])
streamwater_data
```

```python
# normalize the streamwater data by station
streamwater_data = streamwater_data.astype(float)
streamwater_data = (streamwater_data - streamwater_data.mean()) / streamwater_data.std()
streamwater_data
```

### Load the rainfall data

```python
# load in the rainfall data
rainfall_data = np.load("data/weather_data/rainfall_data.npy")
rainfall_data = rainfall_data.astype(float)
rainfall_data = (rainfall_data - rainfall_data.mean()) / rainfall_data.std()
rainfall_data.shape
```

```python
# load the snowfall data
snowfall_data = np.load("data/weather_data/snowfall_data.npy")
snowfall_data = snowfall_data.astype(float)
snowfall_data = (snowfall_data - snowfall_data.mean()) / snowfall_data.std()
snowfall_data.shape
```

```python
# load the max temperature data
max_temp_data = np.load("data/weather_data/max_temp_data.npy")
max_temp_data = max_temp_data.astype(float)
max_temp_data = (max_temp_data - max_temp_data.mean()) / max_temp_data.std()
max_temp_data.shape
```

```python
use_snow = False

# Parameters
time_window = 30  # Number of days to look back
if use_snow:
    grid_shape = (*rainfall_data.shape[1:], 2)  # Latitude, longitude grid
else:
    grid_shape = (*rainfall_data.shape[1:], 1)  # Latitude, longitude grid
n_stations = streamwater_data.shape[1]  # Streamflow prediction stations

rain_data = np.array([rainfall_data[i-time_window:i] for i in range(time_window, len(rainfall_data))])
stream_data = np.array([streamwater_data.iloc[i-time_window:i] for i in range(time_window, len(streamwater_data))])[:-1]

if use_snow:
    snow_data = np.array([snowfall_data[i-time_window:i] for i in range(time_window, len(snowfall_data))])
    # add the rain and snow data together as separate channels
    x_data = np.stack([rain_data, snow_data], axis=-1)
else:
    x_data = rain_data
y_data = np.array([streamwater_data.iloc[i] for i in range(time_window, len(streamwater_data))])

# shift the x and y data to match with the next day
x_data = x_data[:-1]
y_data = y_data[1:]

# reshape x data to be 3D
x_data = x_data.reshape(-1, time_window, *grid_shape)

x_data.shape, y_data.shape, stream_data.shape
```

### Create the model

```python
import tensorflow as tf
from tensorflow.keras import layers, models

# Model Definition
def create_model():
    # CNN Input: Daily rainfall data (latitude x longitude grid)
    cnn_input = layers.Input(shape=(time_window, *grid_shape), name="Rainfall_Input")
    
    # CNN applied per day (TimeDistributed)
    cnn = layers.TimeDistributed(layers.Conv2D(8, (3, 3), activation='relu', padding='same'))(cnn_input)
    # cnn = layers.TimeDistributed(layers.Conv2D(16, (3, 3), activation='relu', padding='same'))(cnn)
    # cnn = layers.TimeDistributed(layers.MaxPooling2D((2, 2), padding='valid'))(cnn)
    # cnn = layers.TimeDistributed(layers.Conv2D(8, (3, 3), activation='relu', padding='same'))(cnn)
    cnn = layers.TimeDistributed(layers.MaxPooling2D((2, 2), padding='valid'))(cnn)
    cnn = layers.TimeDistributed(layers.Flatten())(cnn)
    
    # LSTM to learn temporal patterns from CNN-extracted features
    lstm = layers.LSTM(64, return_sequences=False, dropout=0.3, recurrent_dropout=0.3, name="LSTM_Layer")(cnn)
    
    # Fully Connected Layer to predict streamflow
    dense = layers.Dense(n_stations, activation='relu')(lstm)
    output = layers.Dense(n_stations, activation='linear', name="Streamflow_Output")(dense)
    
    # Create the model
    model = models.Model(inputs=cnn_input, outputs=output)
    return model

# Instantiate and compile the model
model = create_model()
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
              loss='mse',  # Mean squared error for regression
              metrics=['mae'])  # Mean absolute error for evaluation

# Model Summary
model.summary()
```

```python
# Train/test split
train_frac = 0.7
val_frac = 0.15
test_frac = 0.15

n_train = int(train_frac * len(x_data))
n_val = int(val_frac * len(x_data))
n_test = len(x_data) - n_train - n_val

x_train, y_train = x_data[:n_train], y_data[:n_train]
x_val, y_val = x_data[n_train:n_train+n_val], y_data[n_train:n_train+n_val]
x_test, y_test = x_data[-n_test:], y_data[-n_test:]

# Train the model
history = model.fit(x_train, y_train,
                    validation_data=(x_val, y_val),
                    epochs=8,
                    batch_size=8,
                    verbose=1)

# Evaluate the model
test_loss, test_mae = model.evaluate(x_val, y_val)
print(f"Test Loss: {test_loss}, Test MAE: {test_mae}")
```

```python
# plot the training and validation loss
import matplotlib.pyplot as plt

plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
# plot the model predictions for the first 20 stations using subplots
n_plot = 20

predictions = model.predict(x_test)

fig, axs = plt.subplots(n_plot // 2, 2, figsize=(16, n_plot * 2))
for i in range(n_plot):
    mse = np.mean((y_test[:, i] - predictions[:, i]) ** 2)
    station = i
    ax = axs[i // 2, i % 2]
    ax.plot(y_test[:, station], label='Actual')
    ax.plot(predictions[:, station], label='Predicted')
    ax.set_title(f"Station {station} (MSE: {mse:.3f})")
    ax.legend()
plt.tight_layout()
plt.show()
```

### Incorporate streamwater data

```python
import tensorflow as tf
from tensorflow.keras import layers, models, regularizers

# Model Definition
def create_model():
    # CNN Input: Daily rainfall data (latitude x longitude grid)
    rainfall_input = layers.Input(shape=(time_window, *grid_shape), name="Rainfall_Input")
    
    # CNN applied per day (TimeDistributed)
    cnn = layers.TimeDistributed(layers.Conv2D(8, (3, 3), activation='relu', padding='same'))(rainfall_input)
    cnn = layers.TimeDistributed(layers.MaxPooling2D((2, 2), padding='valid'))(cnn)
    cnn = layers.TimeDistributed(layers.Flatten())(cnn)
    
    # LSTM to learn temporal patterns from CNN-extracted features
    rainfall_features = layers.LSTM(64, return_sequences=False, name="Rainfall_features")(cnn)
    
    # Streamflow Input: Daily streamflow data (n_stations)
    streamflow_input = layers.Input(shape=(time_window, n_stations), name="Streamflow_Input")
    streamflow_features = layers.LSTM(64, return_sequences=False, name="Streamflow_LSTM")(streamflow_input)
    
    # Merge features from both paths
    merged = layers.Concatenate(name="Merge_Layers")([rainfall_features, streamflow_features])
    
    # Fully connected layers for prediction
    dense = layers.Dense(128, activation='relu', kernel_regularizer=regularizers.l2(0.001))(merged)
    dense = layers.Dropout(0.3)(dense)
    output = layers.Dense(n_stations, activation='linear', name="Streamflow_Output")(dense)
    
    # Create and compile the model
    model = models.Model(inputs=[rainfall_input, streamflow_input], outputs=output)
    return model

# Instantiate and compile the model
model = create_model()
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
              loss='mse',  # Mean squared error for regression
              metrics=['mae'])  # Mean absolute error for evaluation

# Model Summary
model.summary()
```

```python
# split up the stations

def create_model():
    # Rainfall Input Path (CNN + LSTM)
    rainfall_input = layers.Input(shape=(time_window, *grid_shape), name="Rainfall_Input")
    cnn = layers.TimeDistributed(layers.Conv2D(8, (3, 3), activation='relu', padding='same'))(rainfall_input)
    cnn = layers.TimeDistributed(layers.MaxPooling2D((2, 2)))(cnn)
    cnn = layers.TimeDistributed(layers.GlobalAveragePooling2D())(cnn)
    rainfall_features = layers.LSTM(64, return_sequences=False, name="Rainfall_LSTM")(cnn)
    
    # Streamflow Input Path (Per-Station LSTMs)
    streamflow_input = layers.Input(shape=(time_window, n_stations), name="Streamflow_Input")
    per_station_lstm = layers.TimeDistributed(
        layers.LSTM(16, return_sequences=False),
        name="Per_Station_LSTM"
    )(tf.expand_dims(streamflow_input, axis=-1))
    
    # Combine Features
    merged = layers.Concatenate(name="Merge_Layers")([rainfall_features, per_station_lstm])
    
    # Split features for each station and apply dense layers independently
    station_outputs = []
    for station in range(n_stations):
        # Extract features for each station
        station_feature = layers.Lambda(lambda x: x[:, station])(merged)
        
        # Dense layers specific to each station
        station_dense = layers.Dense(32, activation='relu', kernel_regularizer=regularizers.l2(0.001))(station_feature)
        station_dense = layers.Dropout(0.3)(station_dense)
        station_output = layers.Dense(1, activation='linear', name=f"Station_{station}_Output")(station_dense)
        station_outputs.append(station_output)
    
    # Combine all station outputs
    final_output = layers.Concatenate(name="Final_Output")(station_outputs)
    
    # Create and compile the model
    model = models.Model(inputs=[rainfall_input, streamflow_input], outputs=final_output)
    return model
```

```python
# Train/test split
train_frac = 0.7
val_frac = 0.15
test_frac = 0.15

n_train = int(train_frac * len(x_data))
n_val = int(val_frac * len(x_data))
n_test = len(x_data) - n_train - n_val

x_train, y_train = x_data[:n_train], y_data[:n_train]
x_val, y_val = x_data[n_train:n_train+n_val], y_data[n_train:n_train+n_val]
x_test, y_test = x_data[-n_test:], y_data[-n_test:]
stream_train = stream_data[:n_train]
stream_val = stream_data[n_train:n_train+n_val]
stream_test = stream_data[-n_test:]

# Train the model
history = model.fit([x_train, stream_train], y_train,
                    validation_data=([x_val, stream_val], y_val),
                    epochs=12,
                    batch_size=8,
                    verbose=1)

# Evaluate the model
test_loss, test_mae = model.evaluate([x_val, stream_val], y_val)
print(f"Test Loss: {test_loss}, Test MAE: {test_mae}")
```

```python
# plot the model predictions for the first 20 stations using subplots
import matplotlib.pyplot as plt
n_plot = 20

predictions = model.predict([x_val, stream_val])

fig, axs = plt.subplots(n_plot // 2, 2, figsize=(16, n_plot * 2))
for i in range(n_plot):
    mse = np.mean((y_val[:, i] - predictions[:, i]) ** 2)
    station = i
    ax = axs[i // 2, i % 2]
    ax.plot(y_val[:, station], label='Actual')
    ax.plot(predictions[:, station], label='Predicted')
    ax.set_title(f"Station {station} (MSE: {mse:.3f})")
    ax.legend()
plt.tight_layout()
plt.show()
```

```python
def calculate_nse(y_true, y_pred):
    """
    Compute the Nash-Sutcliffe Efficiency (NSE).
    Args:
        y_true: Array of true values (observations).
        y_pred: Array of predicted values.
    Returns:
        NSE value.
    """
    numerator = np.sum((y_true - y_pred) ** 2)
    denominator = np.sum((y_true - np.mean(y_true)) ** 2)
    nse = 1 - numerator / denominator
    return nse

station_nse = [calculate_nse(y_val[:, i], predictions[:, i]) for i in range(y_val.shape[1])]
nse_score = np.mean(station_nse)
print(f"NSE: {nse_score:.4f}")
```
